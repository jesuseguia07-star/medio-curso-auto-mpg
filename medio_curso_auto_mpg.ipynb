# Medio Curso 1 - Parte Práctica
## Análisis de Dataset Auto MPG

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.dtypes
```

```python
df['horsepower'].isnull().sum()
```

```python
mediana_hp = df['horsepower'].median()
df['horsepower'].fillna(mediana_hp, inplace=True)
```

```python
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']
df.head()
```

## Visualización de Datos

```python
plt.figure()
plt.hist(df['acceleration'], bins=20)
plt.title('Distribución de la Aceleración')
plt.xlabel('Aceleración')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure()
plt.scatter(df['displacement'], df['horsepower'])
plt.title('Displacement vs Horsepower')
plt.xlabel('Displacement')
plt.ylabel('Horsepower')
plt.show()
```

```python
promedio_mpg = df.groupby('origin')['mpg'].mean()

plt.figure()
promedio_mpg.plot(kind='bar')
plt.title('MPG Promedio por Origen')
plt.xlabel('Origen')
plt.ylabel('MPG')
plt.show()
```
